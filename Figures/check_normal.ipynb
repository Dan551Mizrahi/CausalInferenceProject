```python
from Figures.results_reading import read_all_true_ates
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
true_ates = read_all_true_ates(50)
list_of_t1_ates = []
list_of_t2_ates = []
for table in true_ates:
    dataframe = pd.DataFrame(table)
    list_of_t1_ates.append(dataframe.loc['T=0', 'T=1'])
    list_of_t2_ates.append(dataframe.loc['T=0', 'T=2'])

# Draw the distribution
plt.hist(list_of_t1_ates, bins=20, alpha=0.5, label='T1')
plt.hist(list_of_t2_ates, bins=20, alpha=0.5, label='T2')
plt.legend(loc='upper right')
plt.show()
```

```python
# QQ plot for T=1
stats.probplot(list_of_t1_ates, dist="norm", plot=plt)
plt.show()
```

```python
# QQ plot for T=2
stats.probplot(list_of_t2_ates, dist="norm", plot=plt)
plt.show()
```
